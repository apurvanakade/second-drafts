```python
def find_cycle(n):
  list = [0, 1]
  current = 1
  while True:
    list.append(current)
    current = (list[-1] + list[-2]) % n
    if current == 0:
      return list

def find_cycle_length(n):
  first = 0
  second = 1
  # current = 1
  counter = 2
  while True:
    current = (first + second) % n
    if current == 0:
      return counter
    first = second
    second = current
    counter +=1
```

```python
# n = 5

ok_nums = []

good_nums = []

very_good_nums = []

for n in range(3, 10000):
  cycle_length = find_cycle_length(n)
  if cycle_length % n == 0:
    ok_nums.append(n)
  if n % cycle_length == 0:
    good_nums.append(n)
  if cycle_length == n:
    very_good_nums.append(n)

print(f"Ok numbers: {ok_nums}")
print(f"Good numbers: {good_nums}")
print(f"Very Good numbers: {very_good_nums}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def fibonacci_cycle(n):
    """
    Find the smallest k such that F_k ≡ 0 (mod n), where F_k is the k-th Fibonacci number.
    This is called the entry point or rank of apparition of n in the Fibonacci sequence.
    
    Args:
        n: The modulus
    
    Returns:
        The smallest k such that n divides F_k
    """
    if n == 1:
        return 1  # F_1 = 1 ≡ 0 (mod 1)
    
    if n == 2:
        return 3  # F_3 = 2 ≡ 0 (mod 2)
    
    # Start with F_0 = 0, F_1 = 1
    prev, curr = 0, 1
    
    # Check F_0 first (but we typically skip k=0)
    # Generate Fibonacci numbers until we find one divisible by n
    for k in range(1, 6 * n + 1):  # Upper bound
        if curr % n == 0:
            return k
        prev, curr = curr, (prev + curr) % n
    
    return -1  # Should never reach here for valid input

def calculate_fibonacci_cycles(max_n=100):
    """
    Calculate the Fibonacci entry points for integers from 1 to max_n.
    
    Args:
        max_n: Maximum value of n to compute
        
    Returns:
        tuple: (n_values, cycles)
    """
    n_values = list(range(1, max_n + 1))
    cycles = [fibonacci_cycle(n) for n in n_values]
    return cycles

def plot_fibonacci_cycles(m, cycles):
    """
    Plot the Fibonacci entry points.
    
    Args:
        n_values: List of n values
        cycles: Corresponding list of entry points
    """
    plt.figure(figsize=(8, 8))
    plt.plot(range(1, m+1), cycles[:m], 'b-', linewidth=1, alpha=0.7)
    plt.scatter(range(1, m+1), cycles[:m], c='red', s=10, alpha=0.6)

    plt.xlabel('n', fontsize=12)
    plt.ylabel('First k such that n | F_k', fontsize=12)
    plt.title('Fibonacci Entry Points: Smallest k where F_k ≡ 0 (mod n)', fontsize=14)
    plt.grid(True, alpha=0.3)

    # add the line y = x
    plt.plot(range(1, m+1), range(1, m+1), 'g--', linewidth=0.5, label='y = x')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_cycle_ratio(m, cycles):
    """
    Plot the ratio of cycles/n for each n.
    
    Args:
        m: Maximum value of n to compute
        cycles: Corresponding list of entry points
    """
    ratios = [ep / n for n, ep in zip(range(1, m + 1), cycles)]
    
    plt.figure(figsize=(8, 8))
    plt.plot(range(1, m+1), ratios, 'b-', linewidth=1, alpha=0.7)
    plt.scatter(range(1, m+1), ratios, c='red', s=10, alpha=0.6)

    plt.xlabel('n', fontsize=12)
    plt.ylabel('α(n) / n', fontsize=12)
    plt.title('Ratio of Fibonacci Entry Point to n', fontsize=14)
    plt.grid(True, alpha=0.3)
    
    # Add horizontal line at y = 1
    plt.axhline(y=1, color='g', linestyle='--', linewidth=0.5, label='y = 1')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return ratios

def find_equal_cycles(n_values, cycles):
    """
    Find all values of n where cycle(n) == n
    
    Args:
        m: Maximum value of n to compute
        cycles: Corresponding list of entry points
        
    Returns:
        List of n values where α(n) = n
    """
    equal_points = [n for n, ep in zip(n_values, cycles) if ep == n]
    return equal_points

n = 10000
m = 1000

# Calculate the entry points
cycles = calculate_fibonacci_cycles(n)

# Plot the entry points
plot_fibonacci_cycles(m, cycles)

# Plot the ratios
ratios = plot_cycle_ratio(m, cycles)

# Find and print all n where α(n) = n
equal_points = find_equal_cycles(range(1, n + 1), cycles)
print(f"Values of n where α(n) = n (up to {n}):")
print(equal_points)
```
